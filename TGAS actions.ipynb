# Actions

Just using those stars with $\Delta\varpi/\varpi<0.2$. Sampling from covariant error distribution fifty times, computing actions, angles and frequencies and finding median and standard deviation of result.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
rave = pd.read_csv('/data/jls/TGAS/rave_actions.csv')
legue = pd.read_csv('/data/jls/TGAS/legue_actions.csv')
apogee = pd.read_csv('/data/jls/TGAS/apogee_actions.csv')
galah = pd.read_csv('/data/jls/TGAS/galah_actions.csv')
data = [rave,legue,apogee,galah]
labels=['RAVE-on','LEGUE','APOGEE','GALAH']
```

```python
hist_args={'bins':40,'lw':2,'histtype':'step','normed':True}
```

```python
for d,l in zip(data,labels):
    x=plt.hist(np.log10(d.JR),range=[-1.,4.],label=l,**hist_args)
plt.xlabel(r'$\log_{10}(J_R/\mathrm{kpc\,km\,s}^{-1})$')
plt.legend()
```

```python
for d,l in zip(data,labels):
    x=plt.hist(np.log10(d.Jz),range=[-2.,3.],label=l,**hist_args)
plt.xlabel(r'$\log_{10}(J_z/\mathrm{kpc\,km\,s}^{-1})$')
plt.legend()
```
