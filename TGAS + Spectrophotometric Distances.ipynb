# Distances
Take all overlap stars between TGAS and RAVE, APOGEE, LEGUE and GALAH separately. Use all those with 2MASS magnitudes to find spectrophotometric distances folded with parallax uncertainty. No cuts on parallax quality.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
rave = pd.read_csv('/data/jls/TGAS/rave_distances_all.csv')
legue = pd.read_csv('/data/jls/TGAS/legue_distances_all.csv')
apogee = pd.read_csv('/data/jls/TGAS/apogee_distances_all.csv')
galah = pd.read_csv('/data/jls/TGAS/galah_distances_all.csv')
data = [rave,legue,apogee,galah]
labels=['RAVE-on','LEGUE','APOGEE','GALAH']
```

```python
hist_args={'bins':40,'lw':2,'histtype':'step','normed':True}
```

```python
for d,l in zip(data,labels):
    x=plt.hist(d.DIST_Padova,range=[0.,2.5],label=l,**hist_args)
plt.xlabel(r'$s/\mathrm{kpc}$')
plt.legend()
```

```python
for d,l in zip(data,labels):
    x=plt.hist(d.J,range=[6.,12.],label=l,**hist_args)
plt.xlabel(r'$K$')
plt.legend(loc=2)
```

```python
for d,l in zip(data,labels):
    x=plt.hist(d.Z_Padova,range=[-1.,0.6],label=l,**hist_args)
plt.xlabel(r'$[\mathrm{Fe}/\mathrm{H}]$')
plt.legend(loc=2)
```

```python
for d,l in zip(data,labels):
    x=plt.hist((d.TGAS_par-d.PAR_Padova)/(.5*d.TGAS_par+.5*d.PAR_ERR_Padova),range=[-1.5,1.5],label=l,**hist_args)
plt.xlabel(r'$\Delta\varpi/\varpi$')
xx=np.linspace(-1.5,1.5)
plt.plot(xx,np.exp(-xx**2/2.)/np.sqrt(2.*np.pi),color='k')
plt.legend(loc=2)
```

```python
for d,l in zip(data,labels):
    x=plt.plot(d.TGAS_par,d.PAR_Padova,'.',ms=5,alpha=1,label=l)
plt.xlabel(r'TGAS parallax/mas')
plt.ylabel(r'SpectroPhotoTGAS parallax/mas')
plt.xlim(0.,10.)
plt.ylim(0.,10.)
```
